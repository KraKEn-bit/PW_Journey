*1. What is a random variable in probability theory?**
A random variable is a numerical outcome of a random phenomenon. It assigns a real number to each outcome in a sample space.

**2. What are the types of random variables?**
There are two main types:

* **Discrete random variables**: Take on a countable number of values.
* **Continuous random variables**: Take on an infinite number of values within a range.

**3. What is the difference between discrete and continuous distributions?**

* **Discrete distributions** describe random variables that take specific, separate values (e.g., number of heads in coin tosses).
* **Continuous distributions** describe variables that can take any value in a range (e.g., height, weight).

**4. What are probability distribution functions (PDF)?**
For **continuous variables**, the **PDF** describes the likelihood of the variable taking on a range of values. It’s a function where the area under the curve between two values gives the probability of the variable falling in that interval.

**5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

* **PDF** shows the probability *density*.
* **CDF** shows the cumulative probability that a variable is less than or equal to a certain value. It’s the integral (area) of the PDF up to that point.

**6. What is a discrete uniform distribution?**
A distribution where all possible outcomes are equally likely. Example: Rolling a fair die (each number from 1 to 6 has a probability of 1/6).

**7. What are the key properties of a Bernoulli distribution?**

* Only two outcomes: success (1) and failure (0).
* A single trial with probability **p** of success and **1 - p** of failure.
* Mean = **p**, Variance = **p(1 - p)**.

**8. What is the binomial distribution, and how is it used in probability?**
The binomial distribution models the number of successes in **n** independent Bernoulli trials with success probability **p**.
Used for scenarios like “how many heads in 10 coin tosses.”

**9. What is the Poisson distribution and where is it applied?**
The Poisson distribution models the number of events occurring in a fixed interval of time or space, assuming events occur independently and at a constant average rate.
Used in fields like call center traffic, radioactive decay, or number of emails per hour.

**10. What is a continuous uniform distribution?**
A distribution where all values within an interval are equally likely. For example, the probability of any point between 0 and 1 in a uniform distribution is the same.

**11. What are the characteristics of a normal distribution?**

* Bell-shaped and symmetric about the mean.
* Mean = Median = Mode.
* Defined by two parameters: mean (μ) and standard deviation (σ).
* Empirical rule: \~68% within 1σ, 95% within 2σ, 99.7% within 3σ.

**12. What is the standard normal distribution, and why is it important?**
It’s a special case of the normal distribution with mean 0 and standard deviation 1.
Important because any normal distribution can be converted to the standard normal using Z-scores.

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
The CLT states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population’s distribution.
This makes it possible to use normal distribution techniques for inference.

**14. How does the Central Limit Theorem relate to the normal distribution?**
It explains why normal distribution appears so often: averages of many samples tend to be normally distributed, even if the data isn’t.

**15. What is the application of Z statistics in hypothesis testing?**
Z statistics are used to determine how many standard deviations a sample statistic is from the population parameter. It helps assess the significance of results in hypothesis testing.

**16. How do you calculate a Z-score, and what does it represent?**
**Z = (X - μ) / σ**, where X is the value, μ is the mean, and σ is the standard deviation.
It represents how far a value is from the mean in terms of standard deviations.

**17. What are point estimates and interval estimates in statistics?**

* **Point estimate**: A single value estimate of a population parameter (e.g., sample mean).
* **Interval estimate**: A range within which the parameter is expected to lie (e.g., confidence interval).

**18. What is the significance of confidence intervals in statistical analysis?**
They give a range of values likely to contain the true population parameter. A 95% confidence interval means we’re 95% confident the interval contains the parameter.

**19. What is the relationship between a Z-score and a confidence interval?**
Z-scores define the critical values that determine the range of a confidence interval in standard normal distributions. For example, ±1.96 Z-scores cover 95% of the distribution.

**20. How are Z-scores used to compare different distributions?**
By converting values from different distributions to Z-scores, we standardize them, allowing for direct comparison regardless of original scales.

**21. What are the assumptions for applying the Central Limit Theorem?**

* The samples are independent.
* The sample size is sufficiently large (typically n ≥ 30).
* The population has a finite variance.

**22. What is the concept of expected value in a probability distribution?**
The expected value (mean) is the long-run average outcome. It's calculated as the sum of all possible values each multiplied by its probability.

**23. How does a probability distribution relate to the expected outcome of a random variable?**
The distribution provides the probabilities of outcomes, and the expected value is the weighted average of those outcomes, showing the typical result over many trials.
