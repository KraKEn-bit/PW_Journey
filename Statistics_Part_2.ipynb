1)What is hypothesis testing in statistics?
Hypothesis testing is a method used to make inferences or decisions about a population based on sample data.

2)What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis (H₀) assumes no effect or no difference; the alternative hypothesis (H₁) suggests there is an effect or difference.

3)What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the probability threshold (commonly 0.05) used to determine whether to reject H₀; it controls the risk of a Type 1 error.

4)What does a P-value represent in hypothesis testing?
The P-value is the probability of obtaining results as extreme as the observed data, assuming H₀ is true.

5)How do you interpret the P-value in hypothesis testing?
If the P-value ≤ α, reject H₀ (evidence supports H₁); if P-value > α, fail to reject H₀.

6)What are Type 1 and Type 2 errors in hypothesis testing?
Type 1 error: rejecting H₀ when it is true. Type 2 error: failing to reject H₀ when H₁ is true.

7)What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
A one-tailed test checks for an effect in one direction; a two-tailed test checks for effects in both directions.

8)What is the Z-test, and when is it used in hypothesis testing?
The Z-test is used when the population variance is known and the sample size is large (n > 30), to compare means or proportions.

9)How do you calculate the Z-score, and what does it represent in hypothesis testing?
Z = (sample mean – population mean) / (standard deviation / √n); it shows how many standard deviations a value is from the mean.

10)What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is used when the sample size is small (n ≤ 30) and population variance is unknown.

11)What is the difference between a Z-test and a T-test?
Z-test uses known population variance; T-test uses sample variance and is for smaller samples.

12)What is the T-test, and how is it used in hypothesis testing?
The T-test compares means when sample size is small and population standard deviation is unknown.

13)What is the relationship between Z-test and T-test in hypothesis testing?
Both test means; T-test is used when standard deviation is unknown and sample size is small; Z-test is for known variance and larger samples.

14)What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval gives a range within which the true population parameter lies with a certain level of confidence (e.g., 95%).

15)What is the margin of error, and how does it affect the confidence interval?
Margin of error is the range added/subtracted from the sample statistic to create the confidence interval; larger margin means wider interval.

16)How is Bayes' Theorem used in statistics, and what is its significance?
Bayes’ Theorem updates the probability of a hypothesis based on new evidence; it’s key in Bayesian inference.

17)What is the Chi-square distribution, and when is it used?
The Chi-square distribution is used for categorical data tests like goodness-of-fit and independence.

18)What is the Chi-square goodness of fit test, and how is it applied?
It tests if a sample matches an expected distribution using observed and expected frequencies.

19)What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is used to compare variances or in ANOVA to assess group differences.

20)What is an ANOVA test, and what are its assumptions?
ANOVA (Analysis of Variance) tests if there are significant differences between group means; assumes normality, independence, and equal variances.

21)What are the different types of ANOVA tests?
One-way ANOVA (one independent variable), Two-way ANOVA (two variables), and Repeated Measures ANOVA.

22)What is the F-test, and how does it relate to hypothesis testing?
The F-test compares variances to test if groups have different means; it is used in ANOVA and regression analysis.

```python
#PRACTICAL-2:

# 1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
from scipy.stats import norm
import numpy as np

sample = np.random.normal(52, 4, 30)
sample_mean = np.mean(sample)
population_mean = 50
population_std = 4

z = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))
print("Z-statistic:", z, "P-value:", p)
print("Result: Reject H0" if p < 0.05 else "Fail to reject H0")

# 2) Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python
sample = np.random.normal(100, 15, 100)
z = (np.mean(sample) - 105) / (np.std(sample) / np.sqrt(len(sample)))
p = 2 * (1 - norm.cdf(abs(z)))
print("Simulated Hypothesis Testing - Z-stat:", z, "P-value:", p)

# 3) Implement a one-sample Z-test using Python to compare the sample mean with the population mean
def one_sample_z_test(data, mu, sigma):
    mean = np.mean(data)
    z = (mean - mu) / (sigma / np.sqrt(len(data)))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

data = np.random.normal(98, 10, 40)
z, p = one_sample_z_test(data, 100, 10)
print("One-sample Z-test: Z =", z, ", P-value =", p)

# 4) Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import matplotlib.pyplot as plt

z_critical = 1.96
x = np.linspace(-4, 4, 500)
y = norm.pdf(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Reject Region')
plt.axvline(z, color='blue', linestyle='--', label=f'z = {z:.2f}')
plt.title("Two-tailed Z-test Decision Region")
plt.legend()
plt.grid()
plt.show()

# 5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
def visualize_type_errors(mu0, mu1, sigma, n, alpha):
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)
    se = sigma / np.sqrt(n)
    z_alpha = norm.ppf(1 - alpha / 2)
    critical_low = mu0 - z_alpha * se
    critical_high = mu0 + z_alpha * se

    plt.plot(x, norm.pdf(x, mu0, se), label="H0: μ=μ0")
    plt.plot(x, norm.pdf(x, mu1, se), label="H1: μ=μ1", linestyle='--')
    plt.axvline(critical_low, color='red', linestyle='--', label="Critical Region")
    plt.axvline(critical_high, color='red', linestyle='--')

    plt.fill_between(x, norm.pdf(x, mu0, se), where=(x < critical_low) | (x > critical_high), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, norm.pdf(x, mu1, se), where=(x >= critical_low) & (x <= critical_high), color='green', alpha=0.3, label="Type II Error")

    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.grid()
    plt.show()

visualize_type_errors(100, 105, 15, 30, 0.05)

# 6) Write a Python program to perform an independent T-test and interpret the results
from scipy.stats import ttest_ind

group1 = np.random.normal(60, 10, 30)
group2 = np.random.normal(65, 10, 30)
t_stat, p_val = ttest_ind(group1, group2)
print("Independent T-test: T =", t_stat, ", P-value =", p_val)

# 7) Perform a paired sample T-test using Python and visualize the comparison results
from scipy.stats import ttest_rel

before = np.random.normal(70, 5, 30)
after = before + np.random.normal(2, 1, 30)
t_stat, p_val = ttest_rel(before, after)
print("Paired T-test: T =", t_stat, ", P-value =", p_val)

plt.plot(before, label="Before")
plt.plot(after, label="After")
plt.title("Paired Sample Comparison")
plt.legend()
plt.grid()
plt.show()

# 8) Simulate data and perform both Z-test and T-test, then compare the results using Python
sample = np.random.normal(50, 10, 20)
mu = 52
z = (np.mean(sample) - mu) / (10 / np.sqrt(len(sample)))
t_stat, p_val = ttest_ind(sample, np.random.normal(mu, 10, 20))

print("Z-test stat:", z)
print("T-test stat:", t_stat, ", P-value:", p_val)

# 9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance
def confidence_interval(data, confidence=0.95):
    mean = np.mean(data)
    se = np.std(data) / np.sqrt(len(data))
    h = norm.ppf((1 + confidence) / 2) * se
    return mean - h, mean + h

sample = np.random.normal(100, 15, 40)
ci = confidence_interval(sample)
print("Confidence Interval (95%):", ci)
print("Interpretation: We are 95% confident the true population mean lies within this interval.")

# 10) Write a Python program to calculate the margin of error for a given confidence level using sample data
def margin_of_error(data, confidence=0.95):
    se = np.std(data) / np.sqrt(len(data))
    return norm.ppf((1 + confidence) / 2) * se

data = np.random.normal(50, 5, 50)
print("Margin of Error (95%):", margin_of_error(data))

# 11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, prior_not_A, likelihood_A, likelihood_not_A):
    numerator = likelihood_A * prior_A
    denominator = numerator + (likelihood_not_A * prior_not_A)
    return numerator / denominator

p = bayes_theorem(0.01, 0.99, 0.9, 0.05)
print("Posterior probability of having the disease given a positive test:", p)

# 12) Perform a Chi-square test for independence between two categorical variables in Python
from scipy.stats import chi2_contingency

observed = np.array([[30, 10], [20, 40]])
chi2, p, dof, expected = chi2_contingency(observed)
print("Chi-square Test: χ² =", chi2, ", P-value =", p)

# 13) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
print("Expected Frequencies:\n", expected)

# 14) Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
observed = np.array([20, 30, 25, 25])
expected = np.array([25, 25, 25, 25])
chi2, p = np.sum((observed - expected)**2 / expected), 1 - norm.cdf(np.sum((observed - expected)**2 / expected))
print("Goodness-of-fit Test: χ² =", chi2, ", P-value ≈", p)
```
